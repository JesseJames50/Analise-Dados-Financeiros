# Outros dados sobre os papéis
## Basta passar o ticker dentro da função.

```python
#Coletando os dados fundamentalistas da Coca-Cola
coke = yf.Ticker('KO')
```

```python
# informações do papel
coke.info
```

```python
# Acessar algum dado das informações
coke.info['bid']
```

```python
# Retorna histórico de dividendos, stock splits, etc
#period = 'max' retorna o histórico desde a existência do papel
hist = coke.history(period = 'max')
```

```python
# Verificando os ultimos registros do Dataframe
hist.tail()
```

```python
# Plotando o gráfico do histórico de fechamentos
hist.Close.plot()
```

```python
# Plotando o gráfico do histórico de dividendos
hist.Dividends.plot()
```

```python
# Verificando o tamanho do arquivo (linhas, colunas)
hist.shape
```

```python
# Plotando o gráfico do histórico de dividendos pelo número de série
hist.Dividends.iloc[14113:15214].plot()
```

```python
# Mostrar eventos de liquidez actions (dividendos, splits)
coke.actions
```
